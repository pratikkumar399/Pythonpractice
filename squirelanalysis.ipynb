```python
gray_color_count = len(df[df['Primary Fur Color'] == 'Gray'])
red_color_count = len(df[df['Primary Fur Color'] == 'Cinnamon'])
black_color_count = len(df[df['Primary Fur Color'] == 'Black'])
```

```python
print(gray_color_count)
print(red_color_count)
print(black_color_count)
```

```python
dict = {
    'Fur Color' : ['gray' ,'cinnamon' ,'black'],
    'count' : [gray_color_count,red_color_count,black_color_count]
}
```

```python
csv_file_df = pd.DataFrame(dict)
csv_file_df.to_csv('squirel_count.csv')
```
