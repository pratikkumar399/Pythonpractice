```python
from io import StringIO, BytesIO
```

```python
data = (
    'col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3'
)
```

```python
type(data)
```

```python
StringIO()
```

```python
pd.read_csv(StringIO(data))
```

```python
# Read from specific column
df=pd.read_csv(StringIO(data), usecols=['col1', 'col3'])
df
```

```python
df.to_csv('Test1.csv')
df
```

```python
## Specifying columns data types

data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11,12')
```

```python
#dtype can be specified exclusively.If the datatype required is int then we can write dtype = int , or else it can be object float etc
df = pd.read_csv(StringIO(data), dtype =int) 
df
```
